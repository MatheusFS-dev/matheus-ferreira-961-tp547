```python
import numpy as np


q=0.08 # Probabilidade
n=20 # Número de tentativas
value=3
N=100000 # Número de amostras
c = q/(1-q)
av=np.array([])
count=0
x=np.random.uniform(0,1,N)
for ix in x:
    i = 0
    pr = pow((1 - q),n)
    F = pr
    while ix>=F:
        pr = (c * (n - i) / (i + 1))* pr;
        F = F + pr;
        i = i + 1;
    a1=i
    av=np.append(av,a1)

print(av)

for binvalue in av:
    if binvalue==value:
        count=count+1
prob=count/N
print("a probabilidade e",prob)
```
